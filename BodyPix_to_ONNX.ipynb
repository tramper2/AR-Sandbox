# BodyPix original model download

Run the following cells to download the original BodyPix models.

```
def download_model(model_name, base_url, manifest_name, shard_count):
  !mkdir "/content/"{model_name}
  %cd "/content/"{model_name}
  !wget {base_url}"/"{manifest_name}
  for i in range(1, shard_count + 1):
    !wget {base_url}"/group1-shard"{i}"of"{shard_count}".bin"
```

```
download_model("MobileNetV1-x050-stride8", "https://storage.googleapis.com/tfjs-models/savedmodel/bodypix/mobilenet/float/050", "model-stride8.json", 1)
download_model("MobileNetV1-x050-stride16", "https://storage.googleapis.com/tfjs-models/savedmodel/bodypix/mobilenet/float/050", "model-stride16.json", 1)
download_model("MobileNetV1-x075-stride8", "https://storage.googleapis.com/tfjs-models/savedmodel/bodypix/mobilenet/float/075", "model-stride8.json", 2)
download_model("MobileNetV1-x075-stride16", "https://storage.googleapis.com/tfjs-models/savedmodel/bodypix/mobilenet/float/075", "model-stride16.json", 2)
download_model("MobileNetV1-x100-stride8", "https://storage.googleapis.com/tfjs-models/savedmodel/bodypix/mobilenet/float/100", "model-stride8.json", 4)
download_model("MobileNetV1-x100-stride16", "https://storage.googleapis.com/tfjs-models/savedmodel/bodypix/mobilenet/float/100", "model-stride16.json", 4)
download_model("ResNet50-stride16", "https://storage.googleapis.com/tfjs-models/savedmodel/bodypix/resnet50/float", "model-stride16.json", 23)
download_model("ResNet50-stride32", "https://storage.googleapis.com/tfjs-models/savedmodel/bodypix/resnet50/float", "model-stride32.json", 23)
```

# TensorFlow.js to ONNX conversion

Run the following cells to convert the downloaded model files.

```
def convert_model(model_name, manifest_name, arch_name):
  %cd "/content/"{model_name}
  !tfjs_graph_converter {manifest_name} {model_name}".pb"
  !python -m tf2onnx.convert --graphdef {model_name}".pb" --inputs="sub_2:0" --outputs="float_segments:0,float_heatmaps:0,float_short_offsets:0" --output "/content/"{model_name}".onnx"
```

```
convert_model("MobileNetV1-x050-stride8", "model-stride8.json", "MobilenetV1")
convert_model("MobileNetV1-x050-stride16", "model-stride16.json", "MobilenetV1")
convert_model("MobileNetV1-x075-stride8", "model-stride8.json", "MobilenetV1")
convert_model("MobileNetV1-x075-stride16", "model-stride16.json", "MobilenetV1")
convert_model("MobileNetV1-x100-stride8", "model-stride8.json", "MobilenetV1")
convert_model("MobileNetV1-x100-stride16", "model-stride16.json", "MobilenetV1")
convert_model("ResNet50-stride16", "model-stride16.json", "resnet_v1_50")
convert_model("ResNet50-stride32", "model-stride32.json", "resnet_v1_50")
```
